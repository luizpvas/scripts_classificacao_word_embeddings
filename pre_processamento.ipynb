```python
import sys
from nltk.corpus import stopwords
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
from gensim.models.wrappers import FastText
```

```python
init_notebook_mode(connected=True)
model = FastText.load_fasttext_format('wiki.simple/wiki.simple')
```

```python
messages = []
with open('./20ng-train-all-terms.txt') as f:
    for line in f.readlines():
        parts = line.split('\t')
        messages.append({
            'topic': parts[0],
            'words': parts[1].split(' ')
        })
```

```python
stopWords = set(stopwords.words('english'))
```

```python
# Remover as palavras com base nas stopwords
for message in messages:
    message['words_no_stop'] = [x for x in message['words'] if x not in stopWords]
```

```python
# Remover as palavras com base na existência ou não existência no espaço vetorial
removidas = []
for message in messages:
    message['words_no_embedding'] = [x for x in message['words_no_stop'] if x in model.wv]
    for w in message['words_no_stop']:
        if w not in model.wv:
            removidas.append(w)
sorted(removidas, key=lambda x: -len(x))
```

```python
total_words = sum([len(x['words']) for x in messages])
total_no_stop = sum([len(x['words_no_stop']) for x in messages])
total_no_embedding = sum([len(x['words_no_embedding']) for x in messages])
```

```python
total = [len(x['words']) for x in messages if len(x['words']) < 1000]
no_stop = [len(x['words_no_stop']) for x in messages if len(x['words_no_stop']) < 1000]

figure = go.Figure(data=[
    go.Histogram(
        x = no_stop,
        opacity = 0.7,
    ),
    go.Histogram(
        x = total,
        opacity = 0.6,
    )
], layout=go.Layout(
    barmode='overlay'
))

iplot(figure)
```
